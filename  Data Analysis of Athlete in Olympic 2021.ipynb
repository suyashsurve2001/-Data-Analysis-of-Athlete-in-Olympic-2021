```python
# Import  Python Libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#load dataset
athletes =pd.read_csv("athlete_events.csv")
regions= pd.read_csv('national_olympic_committee_regions.csv')
```

```python
athletes.head()
```

```python
regions.head()
```

```python
#join the dataframe
athletes_df=athletes.merge(regions,how='left',on ='NOC')
athletes_df.head()
```

```python
athletes_df.shape
```

```python
# rename the columns name
athletes_df.rename(columns={'region':'Region','notes':'Notes'},inplace=True)
```

```python
athletes_df.head()
```

```python
athletes_df.info()
```

```python
# top 10  countries participating

top_ten_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_ten_countries
```

```python
#plot for the top 10 countries
plt.figure(figsize=(12,6))
plt.title('Top 10  Countries Participating')
sns.barplot(x=top_ten_countries.index, y=top_ten_countries,palette='Set2')
```

```python
 # age distribution of the Participants
plt.figure(figsize=(12,6))
plt.title('Age Distribution of the Participants')
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athletes_df.Age, bins= np.arange(10,80,2),color='orange', edgecolor ='white');
```

```python
#  sports played in winter
winter_sports =athletes_df[athletes_df.Season=='Winter'].Sport.unique()
winter_sports
```

```python
#  sports played in summer
summer_sports =athletes_df[athletes_df.Season=='Summer'].Sport.unique()
summer_sports
```

```python
#male and female participants
gender_counts=athletes_df.Sex.value_counts()
gender_counts
```

```python
# pie plot for male and female participants
plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%',startangle=150,shadow=True);
```

```python
# total medals
athletes_df.Medal.value_counts()
```

```python
#total number of female athletes in each olympics
female_participants = athletes_df[(athletes_df.Sex=='F')&(athletes_df.Season=='Summer')][['Sex','Year']]
female_participants=female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
#GOLD medal athletes
goldMedals = athletes_df[(athletes_df.Medal=='Gold')]
goldMedals.head()
```

```python
goldMedals=goldMedals[np.isfinite(goldMedals['Age'])]
goldMedals
```

```python
goldMedals['ID'][goldMedals['Age']>60].count()
```

```python
sporting_event =goldMedals['Sport'][goldMedals['Age']>60]
sporting_event
```

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold medals for Athletes over 60 years')
```

```python
#gold medals from each countery
goldMedals.Region.value_counts().reset_index(name='Medal').head(10)
```

```python
totalGoldMedals=goldMedals.Region.value_counts().reset_index(name='Medal').head(10)
g=sns.catplot(x="index",y="Medal",data=totalGoldMedals,height=8,kind="bar",palette='rocket')
g.despine(left=True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of Medals")
plt.title('Gold Medals per Country')
```
